## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('Task (2) Dataset.csv')
```

### Identify and analyze patterns in the data.

```python
df.isnull().sum()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], kde=True, bins=20, color='Pink')
plt.title('Histogram of Total Traffic')
plt.xlabel('Total Traffic')
plt.ylabel('Frequency')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
nums = df[['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric', 'TravelTimeLive']]
corr = nums.corr()

corr
```
